# Linear Regression 

### Linear Data

```python
df = pd.read_csv("regression_linear_line.csv")
df.head()
```

```python
X = df["x"].to_numpy()
y = df["y"].to_numpy()
```

```python
def initialize_model(iters=100, reg_param=0.001, regularization="L1"):
    if regularization=="L1":
        reg = Lasso(max_iter=iters, alpha=reg_param, warm_start=True)
    elif regularization=="L2":
        reg = Ridge(max_iter=iters, alpha=reg_param, solver='sag')
    return reg
```

```python
def plot_boundary(X, y, df, iters=100, reg_param=0.001, regularization="L1", n=1):
    f = plt.figure()
    reg = initialize_model(iters=n, reg_param=reg_param, regularization=regularization)
    for i in range(0, iters, n):
        plt.cla()
        reg.fit(X.reshape(-1, 1), y)
        theta = reg.coef_
        b = reg.intercept_
        sns.scatterplot(x="x", y="y", data=df)
        plt.plot(X, theta*X + b, 'r', linewidth=2.5)
        plt.pause(0.001)
    plt.show()
        
```

```python
plot_boundary(X, y, df)
```
