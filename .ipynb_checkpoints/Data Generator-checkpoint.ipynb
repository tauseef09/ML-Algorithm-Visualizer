```python
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import truncnorm
```

# Linearly Separated Classification (Uniform)

```python
data = []
x1_l = []
x2_l = []
y_l = []
n_points = 400
```

```python
for i in range(n_points//2):
    x1 = random.uniform(0, 4)
    x2 = random.uniform(0, 4)
    y = 1
    data.append([x1, x2, y])
    x1_l.append(x1)
    x2_l.append(x2)
    y_l.append(y)

for i in range(n_points//2):
    x1 = random.uniform(-1, -4)
    x2 = random.uniform(-1, -4)
    y = -1
    data.append([x1, x2, y])
    x1_l.append(x1)
    x2_l.append(x2)
    y_l.append(y)
```

```python
plt.scatter(x1_l, x2_l, c=y_l)
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.show()
```

# Non-linear Classification (XOR)

```python
data = []
x1_l = []
x2_l = []
y_l = []
n_points = 400
low = 0.4
high = 5
```

```python
for i in range(n_points//4):
    x1 = random.uniform(low, high)
    x2 = random.uniform(low, high)
    y = 1
    data.append([x1, x2, y])
    x1_l.append(x1)
    x2_l.append(x2)
    y_l.append(y)

for i in range(n_points//4):
    x1 = random.uniform(-low, -high)
    x2 = random.uniform(-low, -high)
    y = 1
    data.append([x1, x2, y])
    x1_l.append(x1)
    x2_l.append(x2)
    y_l.append(y)

for i in range(n_points//4):
    x1 = random.uniform(low, high)
    x2 = random.uniform(-low, -high)
    y = -1
    data.append([x1, x2, y])
    x1_l.append(x1)
    x2_l.append(x2)
    y_l.append(y)

for i in range(n_points//4):
    x1 = random.uniform(-low, -high)
    x2 = random.uniform(low, high)
    y = -1
    data.append([x1, x2, y])
    x1_l.append(x1)
    x2_l.append(x2)
    y_l.append(y)
```

```python
plt.scatter(x1_l, x2_l, c=y_l)
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.show()
```
