# Simple line plot

## The first basic examples use randomly generated data

```python
age = np.array([n for n in range(25, 36)])

salary = np.random.randint(low=25000, high=100000, size=len(age))

plt.plot(age, salary)
plt.title('Salary(USD) by age')
plt.xlabel('Age')
plt.ylabel('Salary(USD)')
plt.show  # would be random each time
```

## Multiple plots together

```python
py_salary = np.random.randint(low=30000, high=100000, size=len(age))
plt.plot(age, salary, label='All devs')
plt.plot(age, py_salary, label='Python devs')

plt.title('Salary(USD) by age')
plt.xlabel('Age')
plt.ylabel('Salary(USD)')
plt.legend()
plt.show()
```

## Loading data from CSV

```python
df = pd.read_csv('stackdata.csv', index_col='Respondent', usecols=['Respondent', 'LanguageWorkedWith'])
df = df['LanguageWorkedWith'].dropna().str.split(';')
df.head()
```

```python
from collections import Counter
languageCount = Counter()
for response in df:
    languageCount.update(response)  # counts the responses
    
languages = list()
langUser = list()
languageCount = languageCount.most_common()  # sorts the counter 

for item in languageCount:
    languages.append(item[0])
    langUser.append(item[1])
```

## Horizontal bar plots

```python
plt.style.use('fivethirtyeight')
# we will reverse it as it starts plotting from the bottom
languages.reverse()
langUser.reverse()
plt.barh(languages, langUser, height=0.4)  # y axis value first
plt.title('Popular Programming Languages')
plt.xlabel('Number of users')
plt.show()
```

## Exploding segments

```python
# shadows
plt.pie(langUser[-6:], labels=languages[-6:], wedgeprops ={'edgecolor': 'k'}, shadow=True, explode = [0.1 * (x == 5) for x in range(6)])
plt.title('Top 6 Programming languages')
```

```python
# showing the percent signs
plt.pie(langUser[-6:], labels=languages[-6:], wedgeprops ={'edgecolor': 'k'}, autopct='%1.1f%%')
plt.title('Top 6 Programming languages')
```

# Getting some data

```python
# reading some data from csv
df = pd.read_csv('stackdata.csv', usecols=['Age', 'ConvertedComp', 'LanguageWorkedWith'])
df = df.dropna()
df = df.loc[(df['Age'] >= 16.0) & (df['Age'] <= 50.0)]
df.set_index('Age', inplace=True)
df.head()
```

```python
grp_obj = df.groupby(by='Age')
grp_obj.median().head()
```

```python
ages = list(set(df.index))[:32]
ages.sort()
medianSalary = grp_obj.median().values
medianSalary = medianSalary.reshape(1, len(medianSalary))[0].tolist()[:32]

py = df.loc[df['LanguageWorkedWith'].str.contains('Python')].groupby(by='Age').median().values.tolist()
py_medianSalary = [item for sublist in py for item in sublist][:32]

js = df.loc[df['LanguageWorkedWith'].str.contains('JavaScript')].groupby(by='Age').median().values.tolist()
js_medianSalary = [item for sublist in js for item in sublist][:32]
```

# Filling area of Line plots

```python
plt.plot(ages, py_medianSalary, label='Python Devs', linestyle='-', color='#3b55ff', linewidth=1)
plt.plot(ages, js_medianSalary, label='JS Devs', linestyle='-', color='#a5ad00', linewidth=1)
plt.plot(ages, medianSalary, label='All Devs', color='#f76331', linewidth=2)

plt.xlabel('Age')
plt.ylabel('Salary in USD')
plt.legend()
plt.fill_between(ages, py_medianSalary, alpha=0.10)  # X axis, the plot to be filled and transparency value
plt.grid(False)
plt.show()
```

```python
# conditional threshold
plt.style.use('fivethirtyeight')
overallMedianSalary = df['ConvertedComp'].median()

plt.plot(ages, py_medianSalary, label='Python Devs', linestyle='-', color='#3b55ff', linewidth=1)
plt.plot(ages, js_medianSalary, label='JS Devs', linestyle='-', color='#a5ad00', linewidth=1)
plt.plot(ages, medianSalary, label='All Devs', color='#f76331', linewidth=2)

plt.xlabel('Age')
plt.ylabel('Salary in USD')
plt.fill_between(ages, py_medianSalary, overallMedianSalary, where=py_medianSalary > overallMedianSalary, 
                 alpha=0.15, interpolate=True, label='Above Avg')  # X axis, the plots to be filled in between, condition
plt.fill_between(ages, py_medianSalary, overallMedianSalary, where=py_medianSalary <= overallMedianSalary, 
                 alpha=0.15, interpolate=True, label='Below Avg')
plt.legend()
plt.grid(False)
plt.show()
```

```python
# Another example
plt.style.use('fivethirtyeight')
plt.plot(ages, js_medianSalary, label='JS Devs', linestyle='-', color='#a5ad00', linewidth=1)
plt.plot(ages, medianSalary, label='All Devs', color='#cb7dff', linewidth=2)

plt.xlabel('Age')
plt.ylabel('Salary in USD')
plt.fill_between(ages, js_medianSalary, medianSalary, alpha=0.1, interpolate=True, color='green')
plt.legend()
plt.grid(False)
plt.show()
```

# Subplots

```python
# cleaning up csv data
df = pd.read_csv('stackdata.csv', usecols=['Age', 'ConvertedComp', 'LanguageWorkedWith'])
df.dropna(inplace=True)

# cleanup
filt = ((df['Age'] >= 16.0) & (df['Age'] <= 60.0))
df = df.loc[filt]

# dividing the data into bins
df['Range'] = pd.cut(df['Age'], bins=[n for n in range(16, 71, 10)], right=True, labels=[16, 20, 30, 40, 50])

allDevs = df.groupby(by='Range')['ConvertedComp'].median().values
ages = df['Range'].value_counts().index.categories.tolist()

# python devs salaries
py = df.loc[(df['LanguageWorkedWith'].str.contains('Python')) & (~df['LanguageWorkedWith'].str.contains('JavaScript'))]
py = py.groupby(by='Range')['ConvertedComp'].median().values

# js devs salaries
js = df.loc[(df['LanguageWorkedWith'].str.contains('JavaScript')) & (~df['LanguageWorkedWith'].str.contains('Python'))]
js = js.groupby(by='Range')['ConvertedComp'].median().values
```

```python
%matplotlib inline
plt.style.use('seaborn')
plt.plot(ages, py, label='Python Devs', color='#5076e6', linewidth=2)
plt.plot(ages, js, label='JavaScript Devs', color='#eded45', linewidth=2)
plt.plot(ages, allDevs, label='All Devs', color='#e85d2e', linestyle='--', linewidth=3)

plt.title('Salary Comparison between Languages')
plt.xlabel('Ages')
plt.ylabel('Salary in USD')
plt.legend()
plt.grid(False)
plt.show()
```

#### Some ideas for subplots

- **Figure** : A window that holds our plots.
- **Axis**: Our actual plot itself.

> A figure can hold mutiple *axis*. We use **plt.gcf()** to get current figure and **plt.gca()** to get current axis.

```python
# object oriented approach to create same grid

fig, ax = plt.subplots()  # by default 1 axis

ax.plot(ages, py, label='Python Devs', color='#5076e6', linewidth=2)
ax.plot(ages, js, label='JavaScript Devs', color='#eded45', linewidth=2)
ax.plot(ages, allDevs, label='All Devs', color='#e85d2e', linestyle='--', linewidth=3)

ax.set_title('Salary Comparison between Languages')
ax.set_xlabel('Ages')
ax.set_ylabel('Salary in USD')
ax.legend()
plt.grid(False)
plt.show()
```

```python
# more axis
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)  # by default axis = 1 row and 1 col

ax1.plot(ages, allDevs, label='All Devs', color='#e85d2e', linestyle='--', linewidth=3)
ax2.plot(ages, py, label='Python Devs', color='#5076e6', linewidth=2)
ax2.plot(ages, js, label='JavaScript Devs', color='#eded45', linewidth=2)

ax1.set_title('Median Salary of Different Ages')
ax1.set_ylabel('Salary in USD')
ax1.legend()
ax1.grid(False)

ax2.set_xlabel('Ages')
ax2.set_ylabel('Salary in USD')
ax2.legend()
ax2.grid(False)

plt.show()
```

```python
# we want 2 seperate figures

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, allDevs, label='All Devs', color='#e85d2e', linestyle='--', linewidth=3)
ax2.plot(ages, py, label='Python Devs', color='#5076e6', linewidth=2)
ax2.plot(ages, js, label='JavaScript Devs', color='#eded45', linewidth=2)

ax1.set_title('Median Salary of Different Ages')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Salary in USD')
ax1.legend()
ax1.grid(False)

ax2.set_title('Salary Comparison')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Salary in USD')
ax2.legend()
ax2.grid(False)

plt.show()
```
